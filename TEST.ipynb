```python
import pandas as pd 
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
%matplotlib inline
```

```python
plt.style.use("ggplot")
plt.rcParams["figure.figsize"] = (12,8)
```

```python
df = pd.read_csv("/Users/prajwalchand/Library/Containers/com.microsoft.Excel/Data/Desktop/MapTheSystem/DataSets/Maternal Mortality Ratio.csv")

df['Year']=df['Year'].astype(float)
df['MMR']=df['MMR'].astype(float)
df['ABR']=df['ABR'].astype(float)


df.set_index('Year').plot()

plt.show()

```

```python
sns.heatmap(df.corr(), annot=True)
```

```python
x = df.ABR
y = df.MMR

plt.scatter(x,y)
plt.show()
```

```python
slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.xlabel("Adolescent Birth Rate (births per 1,000 women ages 15-19)")
plt.ylabel("Maternal Mortality Ratio (deaths per 100,000 live births)")
plt.plot(x, mymodel)
plt.show()

print(r)
```
