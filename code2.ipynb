```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

plt.style.use("ggplot")
plt.rcParams["figure.figsize"] = (15,8)
plt.rcParams["figure.autolayout"] = True

x = np.array([80, 146, 44, 81, 81, 101,79])
y1 =np.array([38.8, 21.6, 41.8, 31.5, 36.4, 31.1 , 38.4])


slope, intercept, r, p, std_err = stats.linregress(x, y1)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.xlabel("Adolescent birth rate aged 15-19 years per 1,000 women in that age group",fontdict={'fontsize': 12})
plt.ylabel('''Demand for family planning satisfied by modern methods, all women aged 15-24:
The proportion of all women aged 15 to 24 who have their 
need for family planning satisfied with modern methods
''', fontdict={'fontsize': 11})

plt.scatter(x,y1)

plt.plot(x, mymodel)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

plt.style.use("ggplot")
plt.rcParams["figure.figsize"] = (15,8)
plt.rcParams["figure.autolayout"] = True

x = np.array([80, 146, 44, 81, 81, 101,79])
y2 =np.array([40.1, 42.2, 49.2, 37.3, 38.9, 44.5, 48.1])


slope, intercept, r, p, std_err = stats.linregress(x, y2)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.xlabel("Adolescent birth rate aged 15-19 years per 1,000 women in that age group",fontdict={'fontsize': 12})
plt.ylabel('''Contraceptive prevalence rate, modern methods, married or in union women:
Percentage of currently married or in union women aged 15 to 49 who are currently using, 
or whose sexual partner is using, at least one modern method of contraception, 
regardless of the modern method used
''', fontdict={'fontsize': 11})

plt.scatter(x,y2)

plt.plot(x, mymodel)
plt.show()
```
